<div style="text-align: center;">
  <h2>Final settings for training of the regression model </h2>
</div>


As we could see in the world map graph, the continent in which the country is located can influence its happiness score, that is why we decided to use the continent in dummie format so that the model can understand it.

```python
concatenated_df = pd.get_dummies(concatenated_df, columns=['Continent'] ,dtype=int)
concatenated_df.drop(columns=['Country', 'Happiness_Rank'], axis=1, inplace=True)
concatenated_df.head(2)
```

Renaming continent columns 

```python
new_columns = {
    'Continent_North America' : 'Continent_North_America',
    'Continent_South America' : 'Continent_South_America'
}

concatenated_df.rename(columns=new_columns, inplace=True)
```

In regression analysis, interaction terms allow us to explore how the relationship between variables impacts the dependent variable. By including interactions, we can capture more complex relationships and improve the model's explanatory power.


```python
concatenated_df['Economy_Health'] = concatenated_df['Economy'] * concatenated_df['Health']

concatenated_df['Trust_Freedom'] = concatenated_df['Trust'] * concatenated_df['Freedom']

concatenated_df['Economy_Trust'] = concatenated_df['Economy'] * concatenated_df['Trust']

concatenated_df['Trust_Health'] = concatenated_df['Trust'] * concatenated_df['Health']

column_order = ['Social_Support', 'Year', 'Trust', 'Generosity','Health', 'Economy', 'Freedom', 'Happiness_Score', 'Continent_Africa', 'Continent_Asia', 'Continent_Europe', 'Continent_North_America', 'Continent_Oceania', 'Continent_South_America', 'Economy_Health', 'Trust_Freedom', 'Economy_Trust','Trust_Health']
concatenated_df = concatenated_df[column_order]

concatenated_df
```

This will be our final daset for use into the machine learning model

---
<div style="text-align: center;">
  <h2>Train and test split</h2>
</div>


By partitioning the data, we create a robust framework for model training and validation, helping to prevent overfitting and ensuring the model's effectiveness on new data. In this case the partitioning will be 70% of the data for training and 30% of the data for test

```python
X = concatenated_df.drop('Happiness_Score', axis=1)
y = concatenated_df['Happiness_Score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=52)
```

```python
print(f'Trainig set\n' + '-'*20)
print(f"Number of Rows: {X_train.shape[0]}\nNumber of Columns : {X_train.shape[1]}")
```

```python
print(f'\nTesting set\n' + '-'*20)
print(f"Number of Rows: {X_test.shape[0]}\nNumber of Columns : {X_test.shape[1]}")
```

---


<div style="text-align: center;">
  <h2>Regression Models </h2>
</div>

We're looking for the best model based on the metric `r^2`

- Linear regression

```python
LR = LinearRegression()
LR.fit(X_train, y_train)
y_pred = LR.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f'R2 Score: {r2}')
```

- Random Forest Regressor 

```python
rf_reg = RandomForestRegressor(n_estimators=150, 
                               random_state=52,
                               )
rf_reg.fit(X_train, y_train)

y_pred = rf_reg.predict(X_test)

r2 = r2_score(y_test, y_pred)
print(f'R2 Score: {r2}')
```

This model is promising but we can seek to adjust the parameters that maximize the performance of the model.

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10]
}

random_forest = RandomForestRegressor(random_state=80)

grid_search = GridSearchCV(random_forest, param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Mejores hiperparmetros:", best_params)

best_model = grid_search.best_estimator_

r2_best_model = best_model.score(X_test, y_test)
print(f'R2 Score del mejor modelo: {r2_best_model}')
```

- Bayesian Ridge

```python
bayesian_ridge = BayesianRidge()

bayesian_ridge.fit(X_train, y_train)

y_pred_bayesian_ridge = bayesian_ridge.predict(X_test)

r2_bayesian_ridge = r2_score(y_test, y_pred_bayesian_ridge)
print(f'R2 Score (Bayesian Ridge): {r2_bayesian_ridge}')
```

- Ridge regression

```python
ridge = Ridge(alpha=1.0)  

ridge.fit(X_train, y_train)

y_pred_ridge = ridge.predict(X_test)

r2_ridge = r2_score(y_test, y_pred_ridge)
print(f'R2 Score (Ridge): {r2_ridge}')
```

- Gradient Boosting Regressor

```python
gradient_boosting = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=52)

gradient_boosting.fit(X_train, y_train)

y_pred_gradient_boosting = gradient_boosting.predict(X_test)

r2_gradient_boosting = r2_score(y_test, y_pred_gradient_boosting)
print(f'R2 Score (Gradient Boosting): {r2_gradient_boosting}')
```

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}

gradient_boosting = GradientBoostingRegressor(random_state=80)

grid_search = GridSearchCV(gradient_boosting, param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Mejores hiperparmetros:", best_params)

best_model = grid_search.best_estimator_

r2_best_model = best_model.score(X_test, y_test)
print(f'R2 Score del mejor modelo: {r2_best_model}')
```
